```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Logistic Regression', 'Neural Network', 'TensorFlow', 'Random Forest', 'Decision Tree', 'K-Nearest Neighbors', 'Gradient Boosting', 'SVM', 'Extra Trees']
accuracies = [84, 80, 77, 90, 88, 73, 89, 85, 91]

plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracies, color='lightgreen', edgecolor='black', linewidth=1.2)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{yval}%', ha='center', va='bottom')

plt.xticks(rotation=45, ha='right')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracies for Dataset 6')

plt.ylim(0, 120)

current_ticks = plt.yticks()[0]
new_ticks = [tick for tick in current_ticks if tick < 120]
plt.yticks(new_ticks)

plt.show()
```
